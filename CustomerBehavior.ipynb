```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_uncleaned = pd.read_csv('Starbucks satisfactory survey.csv')
df_uncleaned.head()
```

```python
df_cleaned = pd.read_csv('Starbucks satisfactory survey encode cleaned.csv')
df_cleaned.head()
```

```python
df_cleaned.describe(include='all')
```

```python
loyal_counts = df_cleaned['loyal'].value_counts()

loyal_counts.index = ['loyal' if i ==1 else 'Not loyal' for i in loyal_counts.index]

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(loyal_counts, labels=loyal_counts.index, autopct='%1.2f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Distribution of loyal')
plt.show()
```

```python
def pieChart(dataframe):
    loyal_counts = dataframe['loyal'].value_counts()
    
    loyal_counts.index = ['loyal' if i ==1 else 'Not loyal' for i in loyal_counts.index]
    
    # Create a pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(loyal_counts, labels=loyal_counts.index, autopct='%1.2f%%', startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.title('Distribution of loyal')
    plt.show()
    
    # print(f"The number of loyal customers: {loyal_counts[1]}")
    # print(f"The number of unloyal customers: {loyal_counts[0]}")
```

```python
df_male = df_cleaned[df_cleaned["gender"] == 0]
df_female = df_cleaned[df_cleaned["gender"] == 1]
```

```python
print("Distribution of male customers")
pieChart(df_male)

print("\n\n\nDistribution of female customers")
pieChart(df_female)
```

### from this it is clear that female customers are more loyal than male

```python
filtered_df = df_cleaned[df_cleaned['loyal'] == 1]

def barGraph(col):
    # Count the occurrences of each income value
    col_counts = filtered_df[col].value_counts().sort_index()
    
    # Values and corresponding counts
    x_values = col_counts.index.tolist()
    y_values = col_counts.values.tolist()
    
    # Create bar plot
    plt.figure(figsize=(8, 6))
    bars = plt.bar(x_values, y_values, color='green', alpha=0.7)
    
    # Add count labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 2), ha='center', va='bottom')
    
    # Set titles and labels
    plt.title(f'Count of {col} Values with Label "1"')
    plt.xlabel(f'{col} Values')
    plt.ylabel('Count with Label "1"')
    plt.xticks(x_values) 
    
    plt.show()
```

```python
# age: 0: Below 20, 1: From 20 to 29, 2: From 30 to 39, 3: 40 and above
barGraph('age')

# 0: student, 1: self-employed, 2: employed, 3: housewife
barGraph('status')

# income: 0: Less than RM25,000, 1 RM25,000 – RM50,000, 2 RM50,000 – RM100,000, 3 RM100,000 – RM150,000, 4 More than RM150,000
barGraph('income')

# visitNo: 0: Daily, 1: Weekly, 3: Monthly, 4: Never
barGraph('visitNo')

# method: 0: Dine In, 1: Drive-thru, 2: Take away, 3: Never, 4: Others
barGraph('method')

# timeSpend: 0: Below 30 mins, 1: 30 mins to 1h, 2: 1h to 2h, 3: 2h to 3 h, 4: More than 3h
barGraph('timeSpend')

# location: 0: Within 1km, 1: 1km to 3km, 2: More than 3km
barGraph('location')
```

### This shows that people aged 20 to 29 is more loyal than other age groups

### This shows that the students and the employed are loyal whereas self employed and housewife are not loyal

### This shows that the loyal customers earn less than RM 25,000

### This shows that people who vist Monthly are more loyal

### This shows that people to Dine in and take away are more loyal

### This shows that people who spend less than an hour are loyal

### This shows that people who live 3km away from the store are more loyal

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = df_cleaned.drop(columns='loyal')
y = df_cleaned['loyal']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
logreg = LogisticRegression(max_iter=400)
logreg.fit(X_train,y_train)
```

```python
(logreg.score(X_test, y_test) * 100 ).round(2)
```
